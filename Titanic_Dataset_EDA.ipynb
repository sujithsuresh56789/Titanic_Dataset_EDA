```python
def imputeage(cols):
    age = cols[0]
    pclass = cols[1]

    if pd.isnull(age):

        if pclass == 1:
            return 37

        elif pclass == 2:
            return 29

        else:
            return 24

    else:
        return age
```

```python
titanic["age"]=titanic[["age","pclass"]].apply(imputeage,axis=1)
```

Balancing data

```python
titanic_df["survived"].value_counts()
```

```python
notsurvived=titanic_df[titanic_df["survived"]==0]
survived=titanic_df[titanic_df["survived"]==1]
```

Under Sampling

```python
notsurvivedsampled=notsurvived.sample(n=340)
```

```python
titanic_df_new=pd.concat([notsurvivedsampled,survived],axis=0)
```

Splitting data into train test split

```python
X=titanic_df_new.drop(["survived"],axis=1)
Y=titanic_df_new["survived"]
```

```python
print(X.shape,Y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,Y_train)
```

```python
totest=logistic.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(Y_test,totest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(Y_test,totest))
```

```python
print(classification_report(Y_test,totest))
```
